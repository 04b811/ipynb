Nota-se que 'Neutral', 'Slightly Positive', e 'Very Positive' tem o valor 0 agora pois não são relevantes para esse estudo (que lida só com como a saúde mental é impactada negativamente)

Soma dos scores de diferentes aspectos de saúde mental, iremos criar colunas de score individuais para cada uma delas:

```python
ADHD = ['ADHD Q1', 'ADHD Q2', 'ADHD Q3', 'ADHD Q4']
dataset['ADHD Score'] = dataset[ADHD].sum(axis=1)

Anxiety = ['Anxiety Q1', 'Anxiety Q2']
dataset['Anxiety Score'] = dataset[Anxiety].sum(axis=1)

SelfEsteem = ['Self Esteem Q1', 'Self Esteem Q2','Self Esteem Q3']
dataset['Self Esteem Score'] = dataset[SelfEsteem].sum(axis=1)

Depression = ['Depression Q1', 'Depression Q2','Depression Q3']
dataset['Depression Score'] = dataset[Depression].sum(axis=1)

Total = ['ADHD Score', 'Anxiety Score','Self Esteem Score','Depression Score']
dataset['Total Score'] = dataset[Total].sum(axis=1)
```

Deletando colunas relacionadas a timestamp já que não são mais necessárias:

```python
dataset.drop(dataset.iloc[:, 9:21], inplace = True, axis = 1)
dataset.drop(['Timestamp'], inplace = True, axis = 1)
```

Iremos adicionar uma coluna de "outcome" (resultado) para evaluar se é recomendado para o indivíduo buscar ajuda psicológica, o dataset pré-define que se a pessoa passar de 40 pontos no score total ela possívelmente está experenciando severos sintomas negativo relacionados a saúde mental (retornar outcome 1), caso contrário ele retornará o outcome 0

```python
def map_score(score):
  if score < 40:
    return "0"
  elif score >= 40:
    return "1"

dataset['Outcome']= dataset['Total Score'].apply(lambda score: map_score(score))
dataset['Outcome'] = dataset['Outcome'].astype('int64')
```
